<h2>O que é música?</h2>

A música é uma forma de arte que se constitui na combinação de vários sons e silêncios, de maneira ordenada e organizada ao longo do tempo.<br/>

É considerada por diversos autores como uma prática cultural e humana. Não se conhece nenhuma civilização ou agrupamento que não possua manifestações musicais próprias. Embora nem sempre seja feito com esse objetivo, a música pode ser considerada como uma forma de arte, considerada por muitos como sua principal função. <br/>

A criação, a performance, o significado e até mesmo a definição de música variam de acordo com a cultura e o contexto social. A música vai desde composições fortemente organizadas (e a sua recriação na performance), música improvisada até formas aleatórias. Pode ser dividida em gêneros e subgêneros, contudo as linhas divisórias e as relações entre gêneros musicais são muitas vezes sutis, algumas vezes abertas à intepretação individual e ocasionalmente controversas. Dentro das "artes", a música pode ser classificada como uma arte de representação, um arte sublime, uma arte de espetáculo.<br/>

Há evidências de que a música é conhecida e praticada desde a pré-história. Provavelmente a observação dos sons da natureza tenha despertado no homem, através do sentido auditivo, a necessidade ou vontade de uma atividade que se baseasse na organização de sons. Embora nenhum critério científico permita estabelecer seu desenvolvimento de forma precisa, a história da música confunde-se com a própria história do desenvolvimento da inteligência e da cultura humana.

Para essa organização de sons que foi evoluindo ao longo dos anos, se dá o nome de **Teoria Musical**. A teoria musical é a forma de explicar como uma música funciona. Ela também é a responsável por mostrar o caminho a ser seguido durante o desenvolvimento musical.


<h2>Introdução a Teoria Musical</h2>

<h3>O que é a teoria musical?</h3>

Antes de saber como funciona a teoria musical, é importante entender o que é o estudo musical teórico.<br/>

De modo geral, teoria musical, ou teoria da música, é um conjunto de normas e sistemas destinado a analisar, classificar, compreender, compor e se comunicar acerca da música. Na prática, ela representa a descrição de todos os elementos musicais e a relação entre a performance prática e a simbologia musical.

Sendo assim, é possível afirmar que a teorial msucai funciona de forma ambígua, sendo tanto algo descritivo como perceptivo. E no final de contas, sua principal característica é facilitar a comunicação entre todos os músicos do mundo através de um sistema único.

Por exemplo: a partir da leitura dos elementos simbólicos que se estabelecem graficamente em uma partitura, diferentes músicos podem decodificar os elementos da música e reproduzi-la usando a voz ou instrumentos como o violão, piano etc.

**Mas o que significa música na teoria musical?**<br/>

Segundo a teoria musical, a música é a combinação de sons e silêncios de uma maneira organizada, sendo composta por melodia, harmonia e ritmo. 

A melodia é a voz principal do som, ou seja, aquilo que pode ser cantado (por uma ou mais vozes).  Já a harmonia é a sobreposição de notas que serve como base para a melodia. Enquanto isso, o ritmo é a marcação do tempo de uma música, que permite ao público acompanhá-la e dançá-la, por exemplo.

---
<h3>Para que serve teoria musical?</h3>

A teoria musical serve para organizar a quantidade de informação que a música carrega como um todo e apresentar os fundamentos essenciais para sua prática. Além dessa organização estrutural, ela também facilita o estudo para iniciantes, oferecendo um modo mais organizado para estudos de acordo com cada escola.

Como todo estudo teórico, a teoria da música também acompanha a sua evolução durante o passar dos anos. Afinal de contas, quanto mais conhecimento sobre diferentes formas de fazer, ler e entender música, mais fértil se torna o terreno para um bom músico evoluir e colocar em prática a criatividade.

<h3>Começando pelo básico</h3>

O primeiro passo para aprender a teoria musical é com os componentes básicos que toda composição musical tem em comum.

Esses são os conceitos desse tópico que você provavelmente conseguirá entender intuitivamente, mesmo que ainda não conheça os seus nomes técnicos.

Os fundamentos básicos são:

* Melodia;
* Harmonia;
* Ritmo.

Com eles em mente, você já poderá passar para os próximos passos da teoria musical.

<h3>Rudimentos da teoria musical</h3>

Os rudimentos são extremamente importantes no momento de estudar este tópico, pois eles também são a base para qualquer músico.

Eles são a teoria musical na prática e estão por trás de qualquer canção tocada no dia a dia, independente do instrumento escolhido.

Os rudimentos são:

* Escalas;
* Acordes;
* Tons;
* Notação.

Cada um deles carrega um assunto bastante completo, porém, com subtópicos muito importantes que precisam ser explorados caso o seu objetivo seja escrever músicas melhores.

<h2>Teoria Musical na Pratica</h2>

<h3>Pauta ou Pentagrama</h3>

Agora vamos as definições e explicações de como a teoria musical funciona na prática. 

A música é a arte de combinar os sons e silêncios de maneira simultânea e sucessivamente, com ordem, eqilíbrio e proporção, dentro do tempo. As principais partes que constituem a música são:
* Melodia: É a combinação dos sons sucessivos (dados uns após outros). É a concepção vertical da música.
* Harmonia: É a combinação dos sons simultâneos (dados de uma só vez). É a concepção vertical da música.
* Ritmo: É a combinação dos valores tempo.

Escreve-se a música sobre 5 linhas e 4 espaços horizontais paralelas e equidistantes, que devem ser contados sempre de baixo para cima. A estas linhas e espaçõs dá-se o nome de PAUTA ou PENTAGRAMA. Veja abaixo a representação de um pentagrama:

<h3>Clave</h3>

A clave é o sinal colocado no início da pauta, sobre determinada linha, para dar nome às notas. As claves são 3 (três):

<img src="assets/01_claves.png"/>

Sequencialmente:
* **Clave de Sol**: Escrito na 2ª linha.
* **Clave de Fá**: É escrita na 3ª ou na 4ª linha.
* **Clave de Dó**: É escrita na 1ª, 2ª, 3ª ou 4ª linha.

<h3>As notas musicais</h3>

Se pensar na música como uma linguagem, então as notas são como letras do alfabeto – Elas são básicas para a construção de uma peça musical. Estudar como o valor das notas se encaixam umas nas outras numa peça musical é ainda mais importante que as suas alturas(pitch) musicais, porque se mudar-mos o valor das notas numa peça de música, acabaremos com uma música completamente diferente.

Existem 7 (sete) notas musicias. O nome das notas variam de acordo com os países, veja:

| A  | B  | C  | D  | E  | F  |   G |
|----|----|----|----|----|----|----:|
| LÁ | SI | DÓ | RÉ | MI | FÁ | SOL |

A segunda notação é comumente utilizada por países cujo idioma oficial origina do Latin, por exemplo: Brasil, Espanha, Itália, etc..
